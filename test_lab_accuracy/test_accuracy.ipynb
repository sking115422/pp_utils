```python
import json
from statistics import mean
```

```python

# Reading in ground truths
gt = open("./ground_truths/result.json")
gt_json = json.load(gt)
gt.close()

# Loading user labels
ul = open("./user_input/result.json")
ul_json = json.load(ul)
ul.close()
```

```python
# Finding only cases that match the gound truth domains

rel_ul_json = []

for a in gt_json:
    gt_domain = a["data"]["image"].split("/")[-1].split("-")[1].split("_")[0]
    for b in ul_json:
        ul_domain = b["data"]["image"].split("/")[-1].split("-")[1].split("_")[0]
        
        if gt_domain == ul_domain:
            
            rel_ul_json.append(b)
```

```python
def calcIOU (x1, y1, w1, h1, xpw1, yph1,  x2, y2, w2, h2, xpw2, yph2):
    
    dx = min(xpw1, xpw2) - max(x1, x2)
    dy = min(yph1, yph2) - max(y1, y2)
    
    aoi = 0
    
    if dx>0 and dy>0:
        aoi = dx*dy
    
    a1 = w1 * h1
    a2 = w2 * h2
    at = a1 + a2
    aou = at-aoi
    
    iou = aoi/(aou + .000000001)
    
    return iou
    
        
    
    
```

```python
calcIOU(1, 1, 5, 5, 6, 6, 2, 3, 8, 7, 10, 10)
```

```python
# Calculating accuracy

for i in range(0, len(gt_json)):
    
    print("")
    print(i, ">", gt_json[i]["data"]["image"].split("/")[-1].split("-")[1])
    print("")
    
    score_list = []
    
    gt_ann_count = len(gt_json[i]["annotations"][0]["result"])
    ul_ann_count = 0
    ind_gt = 0
    ind_match = 0
    
    
    for gt_ann in gt_json[i]["annotations"][0]["result"]:
        
        gt_lab = gt_ann["value"]["rectanglelabels"][0]
        gt_x = float(gt_ann["value"]["x"])
        gt_y = float(gt_ann["value"]["y"])
        gt_w = float(gt_ann["value"]["width"])
        gt_h = float(gt_ann["value"]["height"])
        gt_xpw = gt_x + gt_w
        gt_yph = gt_y + gt_h
        
        ind_gt = ind_gt + 1
        ind_ul = 0
        
        for ul_ann in rel_ul_json[i]["annotations"][0]["result"]:
            ul_lab = ul_ann["value"]["rectanglelabels"][0]
            ul_x = float(ul_ann["value"]["x"])
            ul_y = float(ul_ann["value"]["y"])
            ul_w = float(ul_ann["value"]["width"])
            ul_h = float(ul_ann["value"]["height"])
            ul_xpw = ul_x + ul_w
            ul_yph = ul_y + ul_h
            
            ind_ul = ind_ul + 1
            
            iou = calcIOU(gt_x, gt_y, gt_w, gt_h, gt_xpw, gt_yph, ul_x, ul_y, ul_w, ul_h, ul_xpw, ul_yph)
            
            if iou > 0.50 and gt_lab == ul_lab:
                ind_match = ind_match + 1
                print("  " + str (ind_match) + " > ", gt_lab, iou)
                score_list.append(iou)
                
    ul_ann_count = len(score_list)
    
    print ("")
    if len(score_list) > 0:
        print("    ", "avg element iou: " + str(mean(score_list)))
    print("    ", "ul_count / gt_count: " + str(ul_ann_count/gt_ann_count))
    print("    ", "correct area / total area: " + str(sum(score_list)/gt_ann_count))
    
    
    
                
    
    
                
                    

        
        
```
